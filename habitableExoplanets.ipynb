#### Combining the data

```python
habitable_names = habitableKOI['Name']
habitable_names
```

#### Remove the * character that indicates it is a planetary candidate

```python
habitable_names = habitable_names.tolist()
habitable_names_processed = []
for name in habitable_names:
    if name.endswith('*'):
        habitable_names_processed.append(name[:-1])
    else:
        habitable_names_processed.append(name)
habitable_names_processed
```

```python
KOI['kepler_name'].isin(habitable_names_processed).value_counts()
```

```python
new_habitable_names = []
for name in habitable_names_processed:
    if name.startswith('KOI-'):
        name = name[4:]
        if len(name) == 6:
            name = 'K00' + name
        if len(name) == 7: 
            name = 'K0' + name
        new_habitable_names.append(name)
    else:
        new_habitable_names.append(name)
new_habitable_names
```

```python
KOI['kepler_name'].isin(new_habitable_names).value_counts()
```

```python
KOI['kepoi_name'].isin(new_habitable_names).value_counts()
```

```python
habitableKOI = KOI[(KOI['kepler_name'].isin(new_habitable_names)) | (KOI['kepoi_name'].isin(new_habitable_names))]
habitableKOI.shape
```

```python
habitableKOI.head(10)
```

```python
list(habitableKOI)
```

```python
filteredKOI = habitableKOI.filter(['koi_steff', 'koi_slogg', 'koi_smet', 'koi_srad', 'koi_smass', 'koi_sage'])
filteredKOI.shape
```

```python
filteredKOI.head()
```

```python
filteredKOI['has_habitable'] = 1
filteredKOI.head()
```

```python
filteredKOI['nconfp'] = 1
filteredKOI.describe()
```

```python
filteredKOI = filteredKOI[['koi_steff', 'koi_slogg', 'koi_smet', 'koi_srad', 'koi_smass', 'nconfp', 'has_habitable']]
filteredKOI.head()
```

```python
habitable = filteredKOI
habitable.describe()
```
