#### Column Definitions
 - **kepid** kepler ID
 - **ra** right ascension
 - **dec** declination angle
 - **kepmag** Kepler-band magnitude 
 - **teff** The photospheric temperature of the star
 - **logg** The base-10 logarithm of the acceleration due to gravity at the surface of the star
 - **feh** The base-10 logarithm of the Fe to H ratio at the surface of the star, normalized by the solar Fe to H ratio
 - **radius** The photospheric radius of the star
 - **mass** The mass of the star
 - **dens** The density of the star
 - **nconfp** number of confirmed planets in orbit
 - **nkoi** number of kepler objects of interest
 - **ntce** Number of Kepler TCEs identified in a system
